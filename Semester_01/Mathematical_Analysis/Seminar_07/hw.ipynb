```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plotting e^-x**2
def plot(a, b, n):
      x = np.linspace(a, b, n)
      y = np.exp(-x**2)

      plt.figure(figsize=(10, 6))
      plt.plot(x, y, label=r'$e^{-x^2}$')

      # trapezoidal approx
      for i in range(len(x) -1):
            plt.fill_between(x[i:i+3], y[i:i+3], step="pre", alpha=0.4)

      # plt.fill_between(x, y, alpha=0.3)
      plt.xlabel('x')
      plt.title('Plot of $e^{-x^2}$')
      plt.ylabel(r'$e^{-x^2}$')
      plt.legend()
      plt.grid(True)
      plt.show()

plot(-2, 2, 100)
```

```python
def trapezoidal_approx(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = np.exp(-x**2)
    S = h * (0.5*y[0] + np.sum(y[1:-1]) + 0.5*y[-1])
    return S
```

```python
# change the values here to see how the approximation changes
a = -5
b = 5
n = 400
```

```python
aprox = trapezoidal_approx(a, b, n)
sqrt_pi = np.sqrt(np.pi)

ok = 'OK' if np.abs(aprox - sqrt_pi) < 1e-11 else 'WRONG'
f"Approximation is {aprox:.11f} and the correct value is {sqrt_pi:.11f}. The result is {ok}."
```

```python
plot(a, b, n)
```

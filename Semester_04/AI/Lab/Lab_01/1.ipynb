```python

np.random.seed(44)

t = np.arange(0, 30, 0.01)
nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2

# Two signals with a coherent part at 10 Hz and a random part
s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, NFFT=256, Fs=1. / 0.01)
axs[1].set_ylabel('Coherence')

plt.show()
```

```python
import numpy as np

def spinner(probDistribution):
  probDistribution = np.array(probDistribution)
  if np.sum(probDistribution) == 1:
    return np.random.choice(np.arange(len(probDistribution)), p=probDistribution)
  print(f'Probabilities do not sum to 1')

spinner([0.1, 0.2, 0.3, 0.4])
```

```python
def spinnerDeg(deg):
  deg = np.array(deg)
  if np.sum(deg) == 360:
    probs = deg / np.sum(deg)
    print(probs)
    return np.random.choice(deg, p=probs)
  print(f'Degrees do not sum to 360')

spinnerDeg([45, 85, 70, 160])
```
